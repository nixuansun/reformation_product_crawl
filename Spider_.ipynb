```python


class JCSpider(object):
    def __init__(self):
        # connect to database               you should replace to your own password
        self.db = pymysql.connect(
            host='localhost',
            port=3306,
            user='root',
            passwd='Fx1424sun_',
            db='thereformation',
            charset='utf8'
        )
        self.cursor = self.db.cursor()  
        self.headers = {
            'User-Agent': random.choice(USER_AGENT_LIST),
            'cookie': 'dwanonymous_914668167e2805280fee994f0c25aa1b=ab4xrYazwMXbgqENaIxvmdN4Ug; _gcl_au=1.1.1976538188.1651724172; _ga=GA1.2.1426767983.1651724172; _gid=GA1.2.1945367167.1651724172; rmStore=amid:40090|dmid:8251; GlobalE_CT_Data=%7B%22CUID%22%3A%22149735469.396139115.1008%22%2C%22CHKCUID%22%3Anull%7D; scarab.visitor=%226AA898124CEB1441%22; GlobalE_Welcome_Data=%7B%22showWelcome%22%3Afalse%7D; GlobalE_Full_Redirect=false; tracking_consent=true; __cq_dnt=0; dw_dnt=0; cqcid=ab4xrYazwMXbgqENaIxvmdN4Ug; cquid=||; __cq_uuid=ab4xrYazwMXbgqENaIxvmdN4Ug; dwac_0644a165d42080eb78142cac75=uaV5tsK986OSVX9S0p-oJNq7raNyrcFbsLw%3D|dw-only|||USD|false|US%2FPacific|true; sid=uaV5tsK986OSVX9S0p-oJNq7raNyrcFbsLw; dwsid=k2qOasEFD6zrkwVFPiuXu_B0Wn9xFT1UyXrl2klmnmtMUk5fzOQ0vv7h_DWu-Funzj7TuHJKZup3YIyKkgDClA==; __cq_bc=%7B%22bhcn-reformation-us%22%3A%5B%7B%22id%22%3A%221310298%22%2C%22type%22%3A%22vgroup%22%2C%22alt_id%22%3A%221310298HNA%22%7D%2C%7B%22id%22%3A%221307986%22%2C%22type%22%3A%22vgroup%22%2C%22alt_id%22%3A%221307986LCU%22%7D%5D%7D; __cq_seg=0~0.40!1~0.19!2~0.20!3~0.08!4~-0.21!5~-0.67!6~0.02!7~-0.05!8~0.48!9~0.18; GlobalE_Data=%7B%22countryISO%22%3A%22US%22%2C%22cultureCode%22%3A%22en-US%22%2C%22currencyCode%22%3A%22USD%22%2C%22apiVersion%22%3A%222.1.4%22%7D; _gat_UA-26305999-1=1; _uetsid=1ae00940cc2a11ec9f03611c7d59e588; _uetvid=1ae00760cc2a11ecbcffe9cb68da7c62'
        }
        """Define the visualization window, and set the window and layout"""
        self.window = tk.Tk()
        self.window.title('thereformation data collection')
        self.window.geometry('800x600')

        """create label_user button with manual"""
        self.label_user = tk.Label(self.window, text='Enter the product category to monitor：', font=('Arial', 12), width=30, height=2)
        self.label_user.pack()
        """create label_user related entry"""
        self.entry_user = tk.Entry(self.window, show=None, font=('Arial', 14))
        self.entry_user.pack(after=self.label_user)

        """creat Text box to do demo"""
        # Define a richtextbox slider
        scroll = tk.Scrollbar()
        # Put it on the right side of the window, fill the Y vertical direction
        scroll.pack(side=tk.RIGHT, fill=tk.Y)
        self.text1 = tk.Text(self.window, font=('Arial', 12), width=85, height=20)
        self.text1.pack()
        # Two controls are associated
        scroll.config(command=self.text1.yview)
        self.text1.config(yscrollcommand=scroll.set)

        """Define button 1, bind trigger event method"""
        """click to run and collect the button, and when clicked, the parse_hit_click_1 method will be executed. In real usage scenarios"""
        """parse_hit_click_1 can be replace to real login function, here's only an example """
        self.button_1 = tk.Button(self.window, text='run and collect', font=('Arial', 12), width=10, height=1,
                                  command=self.parse_hit_click_1)
        self.button_1.pack(before=self.text1)

        """Define button 2, bind trigger event method"""
        self.button_2 = tk.Button(self.window, text='remove', font=('Arial', 12), width=10, height=1,
                                  command=self.parse_hit_click_2)
        self.button_2.pack(expand=1)

    def parse_hit_click_2(self):
        """Define trigger event 2, delete the content in the text box"""
        self.text1.delete("1.0", "end")

    def center(self):
        """Create a window centering function method"""
        ws = self.window.winfo_screenwidth()
        hs = self.window.winfo_screenheight()
        x = int((ws / 2) - (800 / 2))
        y = int((hs / 2) - (600 / 2))
        self.window.geometry('{}x{}+{}+{}'.format(800, 600, x, y))

    def run_loop(self):
        """Not able to modify the size of the form"""
        self.window.resizable(False, False)
        """Center the window"""
        self.center()
        """Window maintenance -- persistence"""
        self.window.mainloop()

    def parse_hit_click_1(self):
        """Execute crawler tasks using threads, thread association"""
        Thread(target=self.parse_job_run).start()

    def parse_job_run(self):
        """crawler task function"""
        global keyword
        keyword = self.entry_user.get()
        self.text1.insert("insert", f'Now monitoring {keyword}======================================================' + '\n')
        self.get_link()
        self.text1.insert("insert", f'{keyword}monitor complete' + "\n")
        # self.text1.insert("insert", '\n' + "\n")

    # Get product links for each page
    def get_link(self):
        try:
            global keyword
            hrefs_list = []
            for i in range(10):
                url = f'https://www.thereformation.com/on/demandware.store/Sites-reformation-us-Site/default/Search-UpdateGrid?cgid={keyword}&pmpt=qualifying&srule=Newest&start={i * 30}&sz=30&pageTypeContext=Search-Show&selectedUrl=https://www.thereformation.com/on/demandware.store/Sites-reformation-us-Site/default/Search-UpdateGrid?cgid={keyword}&pmpt=qualifying&srule=Newest&start={i * 30}&sz=30'
                response = session.get(url, headers=self.headers)
                # print(response.status_code)
                if response.status_code == 200:
                    # print(response.text)
                    tree = etree.HTML(response.text)
                    hrefs = tree.xpath('//a[@class="product-tile__anchor"]/@href')[::2]
                    for href in hrefs:
                        href = 'https://www.thereformation.com' + href
                        self.text1.insert("insert", f'{href}' + "\n")
                        self.text1.see('insert')
                        hrefs_list.append(href)
                self.text1.insert("insert", f'Page{i + 1}captured.' + "\n")
            # print(hrefs_list)
            # compare new listed and soldout product with database data
            self.compare(hrefs_list)
        except:
            pass

    def compare(self, hrefs_list):
        try:
            global keyword
            # read MySQL data
            # 
            sql = "select product_url from Reformation where low_level='" + keyword + "'"
            self.cursor.execute(sql)
            results = self.cursor.fetchall()  # 
            # check if there is data in MySQL
            # If not, insert and save directly
            if results == ():
                self.text1.insert("insert", f'The number of newly listed products monitored is：{len(hrefs_list)}' + "\n")
                for append_link in hrefs_list:
                    append_link = append_link.strip()
                    self.text1.insert("insert", f'Links to new products：{append_link}' + "\n")
                    self.text1.see('insert')
                    # Get product detail data
                    self.get_details(append_link)
            # if so，insert and save after comparison
            else:
                old_links = [r[0] for r in results]
                # compare new listed 
                append_links = [x for x in hrefs_list if x not in old_links]
                # compare soldout
                disappear_links = [y for y in old_links if y not in hrefs_list]
                self.text1.insert("insert", f'The number of newly listed products monitored is：{len(append_links)}' + "\n")
                for append_link in append_links:
                    append_link = append_link.strip()
                    self.text1.insert("insert", f'Links to new products：{append_link}' + "\n")
                    self.text1.see('insert')
                    # Get product detail data
                    self.get_details(append_link)

                self.text1.insert("insert", f'The number of soldout product monitored is：{len(disappear_links)}' + "\n")
                for disappear_link in disappear_links:
                    disappear_link = disappear_link.strip()
                    self.text1.insert("insert", f'Links to soldout product：{disappear_link}' + "\n")
                    self.text1.see('insert')
                    self.delete_data(disappear_link)
        except:
            pass

    def get_details(self, append_link):
        try:
            global keyword
            res = session.get(append_link, headers=self.headers).content.decode()
            tree = etree.HTML(res)
            display_name = tree.xpath('//h1[@data-product-component="name"]/text()')[0].strip()
            color = tree.xpath('//span[@class="product-attribute__selected-value"]/text()')[0].strip()
            size = tree.xpath(
                '//div[@data-attr-group="size"]/div[@class="product-attribute__contents flex flex-flow-wrap"]/button/text()')
            price = tree.xpath(
                '//div[@class="pdp__price font-size--22 d-font-size--18"]//span[@class="price--reduced"]/text()')[
                0].strip()
            description = tree.xpath('//div[@class="cms-generic-copy"]/text()')[0].strip()
            brand_name = str(display_name).strip().split(' ')[0]

            params = append_link.split('.html')[0].split('/')[-1][:-3] + '-' + append_link.split('.html')[0].split('/')[
                                                                                   -1][-3:]
            image_links = session.get(f'https://media.thereformation.com/image/list/{params}.json',
                                      headers=self.headers).json()
            public_id = jsonpath(image_links, '$..public_id')
            image_links = [
                'https://media.thereformation.com/image/upload/b_rgb:FFFFFF,c_pad,dpr_2.0,f_auto,h_800,q_auto,w_600/c_pad,h_800,w_600/v1/' + p
                for p in public_id]

            print(display_name)
            print(color)
            print(size)
            print(price)
            print(append_link)
            print(image_links)
            print(brand_name)
            print(description)
            print(today)
            print(keyword)
            print('Women')
            print('False')
            data_list = [display_name, color, str(size), price, append_link, str(image_links), brand_name, description, today,
                         keyword, 'Women', 'False']
            print(data_list)
            # self.text1.insert("insert", f'{data_list}' + "\n")
            print()
            self.save_Reformation(data_list)
            self.text1.insert("insert", f'{display_name}Saved.' + "\n")
        except:
            pass

    def save_Reformation(self, data_list):
        sql = 'insert into Reformation(display_name, color, size, price, product_url, image_links, brand_name, description, scrapped_date, low_level, gender, secondhand) value(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'  # SQL语法
        self.cursor.execute(sql, data_list)  # .execute
        self.db.commit()  # submit

    def delete_data(self, disappear_link):
        sql = "DELETE FROM Reformation where product_url='" + disappear_link + "'"
        self.cursor.execute(sql)  # .execute 
        self.db.commit()  # submit
        self.text1.insert("insert", f'{disappear_link}deleted.' + "\n")


if __name__ == '__main__':
    j = JCSpider()
    j.run_loop()
```
